## Ordinal Number Encoding   --

```python
#
# for using this first we create time ( order wise ) data   --
import pandas as pd
import datetime

today_date = datetime.datetime.today()

# today_date - datetime.timedelta(3)

day = [today_date - datetime.timedelta(x) for x in range(0,15)]

#  now creating dataframe - - -  

data = pd.DataFrame(day)
data.columns = ['Days']

data.head()
```

```python
#
data['Weekdays'] = data['Days'].dt.day_name()

data.head()
```

```python
#  

dictionary = {'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7
    }

data['weekday_count'] = data['Weekdays'].map(dictionary)

data.head(7)
```

###  Probability Ratio Encoding  --

```python
# 

import pandas as pd
import numpy as np

data = pd.read_csv('D:\\MBA\\D A T A     S C I E N C E\\Data\\titanic_train.csv', usecols = ['Cabin', 'Survived'])
data.head()
```

```python
#
data['Cabin'].fillna('Missing', inplace = True)
```

```python
#
data['Cabin'] = data['Cabin'].astype(str).str[0]
data.head()
```

```python
mean = data.groupby(['Cabin'])['Survived'].mean()
mean.head()
```

```python
mean = pd.DataFrame(mean)
mean.head()
```

```python
mean['Dead'] = 1 - mean['Survived']
mean.head()
```

```python
#
prob_ratio = mean['Survived'] / mean['Dead']
prob_ratio.to_dict()
```

```python
#
data['Cabin'] = data['Cabin'].map(prob_ratio)
data.head()
```
